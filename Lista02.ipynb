```python
# ALGORITMO CALCULADORA

print("\n\t\t\t*****************************\n")
print("\t\t\t\t***CLACULADORA***\n")
print("\t\t\t*****************************")

print("===================================================================")
print("(+) para SOMA: ")
print("(-) para SUBTRAÇÃO: ")
print("(*) para MULTIPLICAÇÃO: ")
print("(/) para Divisão: ")
print("===================================================================")

num1 = float(input("Digite um numero: "))
nun2 = float(input("Digite outro numero: "))
equ = str(input("Escolha a opção matemática: "))
resp = float

if equ == '+':
    resp = num1 + nun2
    print("O valor da soma é = ",resp)
else:
    if equ == '-':
        resp = num1 - nun2
        print("O valor da subtração é = ",resp)
    else:
        if equ == '*':
            resp = num1 * nun2
            print("O valor da multiplicação é = ",resp)
        else:
            if equ == '/':
                resp = num1 / nun2
                print("O valor da Divisão é = ",resp)
            else:
                if equ != '+' or equ != '-' or equ != '*' or equ != '/':
                    resp = 'Opção errada!'
                    print(resp)
```
